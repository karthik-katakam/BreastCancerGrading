```python
from glob import glob
import os
import shutil
```

```python
images=sorted(glob("Data/*/images/*"))
masks=sorted(glob("Data/*/masks/*"))

print(len(images))
print(len(masks))
```

```python
source_dir = '../Tubule/Data/T96/'
destination_dir_mask = '../Tubule/Data/T96/masks'
destination_dir_images = '../Tubule/Data/T96/images'

os.makedirs(destination_dir_mask, exist_ok=True)
os.makedirs(destination_dir_images, exist_ok=True)

mask_count = 0
image_count = 0

for filename in os.listdir(source_dir):
    source_path = os.path.join(source_dir, filename)
    
    if "-mask" in filename and filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        destination_path = os.path.join(destination_dir_mask, filename)
        shutil.move(source_path, destination_path)
        print(f"Moved: {filename} to {destination_dir_mask}")
        mask_count += 1  
    elif filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        destination_path = os.path.join(destination_dir_images, filename)
        shutil.move(source_path, destination_path)
        print(f"Moved: {filename} to {destination_dir_images}")
        image_count += 1 

print(f"\nTotal mask images moved: {mask_count}")
print(f"Total non-mask images moved: {image_count}")

```
